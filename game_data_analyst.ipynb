# Step 4: Feature Engineering

## •	Create new features that might be useful for analysis 

### We can add a column that indicates if a game is on a handheld console or not

```python
handheld_consoles = ['DS', 'PSP', 'GBA', '3DS', 'PSV', 'GB', 'GBA', 'WS', 'NG', 'GG']
data['is_handheld'] = data['Platform'].isin(handheld_consoles)
data['is_handheld'] = data['is_handheld'].astype(int)
data[['Platform', 'is_handheld']].value_counts().head(30)
```

```python
region_sales = data.groupby('is_handheld')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
total_sales = region_sales.sum()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

region_sales[['NA_Sales']].plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Sales by Console Type in North America')
axes[0, 0].set_xlabel('Console Type')
axes[0, 0].set_ylabel('Total Sales')
axes[0, 0].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['EU_Sales']].plot(kind='bar', ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Sales by Console Type in Europe')
axes[0, 1].set_xlabel('Console Type')
axes[0, 1].set_ylabel('Total Sales')
axes[0, 1].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['JP_Sales']].plot(kind='bar', ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Sales by Console Type in Japan')
axes[1, 0].set_xlabel('Console Type')
axes[1, 0].set_ylabel('Total Sales')
axes[1, 0].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['Other_Sales']].plot(kind='bar', ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Sales by Console Type in Other Regions')
axes[1, 1].set_xlabel('Console Type')
axes[1, 1].set_ylabel('Total Sales')
axes[1, 1].set_xticklabels(['Home Console', 'Handheld'], rotation=0)


plt.tight_layout()
plt.show()
```

#### This is the popularity of handheld vs home consoles in each region

```python
nintendo_data = data[(data['Publisher'] == 'Nintendo') & (data['Year'].between(2014, 2016))]

region_sales = nintendo_data.groupby('is_handheld')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

region_sales[['NA_Sales']].plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Sales by Console Type in North America \nfor Nintendo Published Games (2014-2016)')
axes[0, 0].set_xlabel('Console Type')
axes[0, 0].set_ylabel('Total Sales')
axes[0, 0].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['EU_Sales']].plot(kind='bar', ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Sales by Console Type in Europe \nfor Nintendo Published Games (2014-2016)')
axes[0, 1].set_xlabel('Console Type')
axes[0, 1].set_ylabel('Total Sales')
axes[0, 1].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['JP_Sales']].plot(kind='bar', ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Sales by Console Type in Japan \nfor Nintendo Published Games (2014-2016)')
axes[1, 0].set_xlabel('Console Type')
axes[1, 0].set_ylabel('Total Sales')
axes[1, 0].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

region_sales[['Other_Sales']].plot(kind='bar', ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Sales by Console Type in Other Regions \nfor Nintendo Published Games (2014-2016)')
axes[1, 1].set_xlabel('Console Type')
axes[1, 1].set_ylabel('Total Sales')
axes[1, 1].set_xticklabels(['Home Console', 'Handheld'], rotation=0)

plt.tight_layout()

plt.show()
```

### As you can see nintendo sales for handheld consoles are increasing. and their last homeconsole was a falilure. so they focused on handheld consoles and made the nintedndo switch

```python
nintendo_data = data[data['Publisher'] == 'Nintendo']
platform_sales = nintendo_data.groupby('Platform')[['Global_Sales']].sum()
platform_sales_sorted = platform_sales.sort_values(by='Global_Sales', ascending=False)

plt.figure(figsize=(10, 9))
platform_sales_sorted.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Nintendo Console')
plt.xlabel('Console')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

-------
Let's say you have sales data in different regions (NA, EU, JP) and you want to create a feature that represents the total sales excluding North America.

```python
data['Sales_Excluding_NA'] = data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales']
data
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Sales_Excluding_NA'], label='Sales Excluding NA', shade=True, color='orange')
sns.kdeplot(data['Global_Sales'], label='Global Sales', shade=True, color='blue')
plt.title('Density Plot of Sales Distributions')
plt.xlabel('Sales (Millions)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The density plot provides a smooth estimate of the distribution of sales, allowing you to compare how the distribution of global sales compares to sales excluding NA. This can highlight the impact of the North American market on the overall sales distribution.

```python
# 3. Line Plot of Yearly Trends
plt.figure(figsize=(14, 6))
yearly_sales_trends = data.groupby('Year')['Sales_Excluding_NA'].sum()
sns.lineplot(x=yearly_sales_trends.index, y=yearly_sales_trends.values, marker='o', color='green')
plt.title('Yearly Trends in Sales Excluding NA')
plt.xlabel('Year')
plt.ylabel('Total Sales Excluding NA (Millions)')
plt.xticks(rotation=45)
plt.show()
```

The line plot of yearly trends will illustrate how sales excluding North America have changed over the years, indicating market growth, decline, or stability outside North America.

```python
# 3. Scatter Plot against Global Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Global_Sales', y='Sales_Excluding_NA', data=data, alpha=0.6)
plt.title('Sales Excluding NA vs. Global Sales')
plt.xlabel('Global Sales (Millions)')
plt.ylabel('Sales Excluding NA (Millions)')
plt.show()
```

The scatter plot helps to understand the relationship between sales excluding NA and the global sales figures, potentially highlighting how significant North American sales are to the global market for different games.
